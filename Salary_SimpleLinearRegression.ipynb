```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df= pd.read_csv('https://raw.githubusercontent.com/ybifoundation/Dataset/main/Salary%20Data.csv')
```

```python
df.head()
```

```python
X_simple = df[['Experience Years']]
y = df['Salary']
simple_model = LinearRegression()
simple_model.fit(X_simple, y)
```

```python
y_pred = simple_model.predict(X_simple)
```

```python
print(y_pred)
```

```python
print(f"Intercept: {simple_model.intercept_}, Coefficient: {simple_model.coef_[0]}")
```

```python
plt.scatter(X_simple, y, color = 'yellow')
plt.plot(X_simple, y_pred, color = 'gray')
plt.title("Simple Linear Regression (Salary, Experience Years)")
plt.xlabel("Experience Years")
plt.ylabel("Salary")
plt.show()
```
