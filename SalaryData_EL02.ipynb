```python
# Salary Data Simple Linear Regression

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
```

```python
df= pd.read_csv('https://raw.githubusercontent.com/SHIELA4/BSIS2A_Felicitas-ShielaMaeB/refs/heads/main/Salary_Data1.csv')
```

```python
df.head()
```

```python
print(df.columns)
```

```python
label_encoder = LabelEncoder()
```

```python
df['Job Title'] = label_encoder.fit_transform(df['Job Title'])
```

```python
X_simple = df[['Job Title']]
y = df['Salary']
```

```python
df = df.dropna(subset=['Salary'])
```

```python
X_simple = df[['Job Title']]
y = df['Salary']
```

```python
simple_model = LinearRegression()
simple_model.fit(X_simple, y)
```

```python
print("Intercept:", simple_model.intercept_)
print("Coefficient:", simple_model.coef_)
```

```python
predictions = simple_model.predict(X_simple)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y, predictions)
r2 = r2_score(y, predictions)
print('Mean Squared Error:', mse)
print('R-squared:', r2)
```

```python
plt.scatter(X_simple, y, color='black')
plt.plot(X_simple, predictions, color='brown')
plt.xlabel('Job Title (Encoded)')
plt.ylabel('Salary')
plt.title('Linear Regression Model')
plt.show()
```
