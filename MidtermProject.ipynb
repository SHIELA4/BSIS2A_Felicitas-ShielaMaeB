```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
# exercise2_SALARY DATA
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
```

```python
sns.histplot(df['Pclass'])
```

```python
plt.show()
```

```python
df1=sns.catplot(data= df,col = "Survived", x = "Pclass", kind="count")
```

```python
df_h=df.groupby(['Survived','Pclass']).size()
```

```python
(df_h.head())
```

```python
top_5=df_h.sort_values(ascending = True).head(5)
```

```python
top_5.plot(kind='bar')
plt.title('Total Pclass')
plt.xlabel('Survived')
plt.ylabel('Highest Pclass')
plt.show()
```

```python
df_h = df.groupby(['Pclass']).size().head(10)
```

```python
print(df_h)
```

```python
df_h.plot(kind='pie')
plt.title('Total Pclass')
plt.show()
```

```python
df_h.plot(kind='line')
plt.title('Highest Pclass')
plt.show()
```

```python
survived_info = df[df['Survived'].notnull()][['Survived']]

#Dropping values of age = 0
survived_info_clean = survived_info[survived_info['Survived'] != 0]
survived_info_clean.info()
```

```python
fig = plt.figure(figsize = (20,8))
Survived_plot = sns.distplot(survived_info_clean, bins = 30)
Survived_plot.tick_params(labelsize = 13)
mean = survived_info['Survived'].mean()

plt.xlabel('Survived', fontdict = {'fontsize':15})
plt.title('Survived passengers', fontdict = {'fontsize':15})
plt.bold=True
plt.axvline(mean,color = 'red')
plt.show()
```
