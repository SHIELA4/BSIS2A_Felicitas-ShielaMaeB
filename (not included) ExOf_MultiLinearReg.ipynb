```python
# Example of Multiple Linear Regression

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/SHIELA4/BSIS2A_Felicitas-ShielaMaeB/refs/heads/main/multiple_linear_regression_dataset.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
X_multi = df[["age", "experience"]]
y = df["income"]
model_multi = LinearRegression()
model_multi.fit(X_multi, y)

print("Intercept:", model_multi.intercept_)
print("Coefficients:", model_multi.coef_)
```

```python
y_pred = model_multi.predict(X_multi)
```

```python
print(y_pred)
```

```python
plt.scatter(X_multi['age'], y, color = 'black')
plt.plot(X_multi['age'], y_pred, color = 'gray')
plt.title("Multiple Linear Regression (experience, income)")
plt.xlabel("income")
plt.ylabel("experience")
plt.show()
```
