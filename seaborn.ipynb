```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/SHIELA4/BSIS2A_Felicitas-ShielaMaeB/main/SALARY_DATA.csv")
```

```python
sns.histplot(df['Salary'])
```

```python
plt.show()
```

```python
df1=sns.catplot(data= df,col = "Name", x = "Salary", kind="count")
```

```python
df_h=df.groupby(['Name','Salary']).size()
```

```python
(df_h.head())
```

```python
top_5=df_h.sort_values(ascending = True).head(5)
```

```python
top_5.plot(kind='bar')
plt.title('Total Salary')
plt.xlabel('Name')
plt.ylabel('Highest Salary')
plt.show()
```

```python
age_info = df[df['Age'].notnull()][['Age']]

#Dropping values of age = 0
age_info_clean = age_info[age_info['Age'] != 0]

age_info_clean.info()
```

```python
fig = plt.figure(figsize = (20,8))
Age_plot = sns.distplot(age_info_clean, bins = 30)
Age_plot.tick_params(labelsize = 13)
mean = age_info['Age'].mean()

plt.xlabel('Age', fontdict = {'fontsize':15})
plt.title('Age of disease-affected humans', fontdict = {'fontsize':15})
plt.bold=True
plt.axvline(mean,color = 'red')
plt.show()
```
